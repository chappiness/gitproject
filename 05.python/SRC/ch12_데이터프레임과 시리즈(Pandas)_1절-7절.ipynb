- get_rdataset(데이터셋명, 패키명)함수로 R 데이터를 가져올 수 있음
    datasets 패키지명은 생략 가능
- get_rdataset()함수로 불러온 데이터는 다음 속성을 가지고 있음
    data : 데이터를 담고 있는 데이터프레임
    __doc__ : 데이터에 대한 설명 문자열
    title : 데이터 이름
